```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("DS_P.xlsx")
pd.set_option('display.max_columns', None)
df.head(15)
```

```python
df=df.drop(["ID","Name","Surname","Customer ID","Unnamed: 9"],axis=1)
```

```python
gender_counts = df['Gender'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
state_counts = df['State'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
state_counts.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.title('Frequency of States in Dataset')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
state_counts = df['Interval'].value_counts()

# Calculate cumulative sum of frequencies
state_cumulative = state_counts.cumsum()

# Calculate percentage of total occurrences
total_occurrences = state_counts.sum()
state_percentage = (state_cumulative / total_occurrences) * 100

# Plotting the Pareto chart
fig, ax1 = plt.subplots()

ax1.bar(state_counts.index, state_counts, color='b')
ax1.set_xlabel('Interval')
ax1.set_ylabel('Frequency', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(state_counts.index, state_percentage, color='r', marker='o', linestyle='-')
ax2.set_ylabel('Cumulative Percentage', color='r')
ax2.tick_params('y', colors='r')

plt.title('Pareto Chart of State Frequencies')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Area (ft.)'], df['Price'])
plt.xlabel('Size')
plt.ylabel('Cost')
plt.title('Size vs Cost (Complexity as Point Size)')
plt.grid(True)
plt.show()
#when we plot a graph between two vairables its better if we find correlation between  the vairables
```
